# Возврат строки

```python
def hello():
    return 'Hello, world!'
```

```python
import dis
dis.dis(hello)
```

```python
dir(hello.__code__)
```

```python
hello.__code__.co_code
```

```python
hello.__code__.co_consts
```

# Возврат целого числа

```python
def myfunc():
    return 5
```

```python
myfunc.__code__.co_consts
```

```python
dis.dis(myfunc)
```

# Возврат списка

# Возврат словаря

```python
def myfunc():
    d = {'a': 1, 'b': 2, 'c': 3}
    return d
```

```python
myfunc.__code__.co_consts
```

```python
myfunc.__code__.co_varnames
```

```python
dis.dis(myfunc)
```

## Случай глобальной переменной

```python
def myfunc():
    global d
    d = {'a': 1, 'b': 2, 'c': 3}
    return d
```

```python
myfunc.__code__.co_varnames
```

```python
dis.dis(myfunc)
```
